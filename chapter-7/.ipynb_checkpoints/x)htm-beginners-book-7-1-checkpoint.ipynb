# Predict Sine Wave

```python
from htm.bindings.sdr import SDR
from htm.algorithms import TemporalMemory as TM
from htm.bindings.encoders import ScalarEncoder, ScalarEncoderParameters

from htm.bindings.algorithms import Predictor

import numpy as np
import matplotlib.pyplot as plt
from IPython.display import display, clear_output
%matplotlib inline

tm = TM(columnDimensions = (1000,),
        cellsPerColumn=1,
        initialPermanence=0.5,
        connectedPermanence=0.5,
        minThreshold=8,
        maxNewSynapseCount=20,
        permanenceIncrement=0.1,
        permanenceDecrement=0.0,
        activationThreshold=8,
        )

predictor = Predictor( steps=[1, 2], alpha=0.1)
predictor_resolution = 1
```

```python
scalarEncoderParams = ScalarEncoderParameters()
scalarEncoderParams.minimum = 0
scalarEncoderParams.maximum = 1
scalarEncoderParams.activeBits = 8
scalarEncoderParams.size = 1000
scalarEncoderParams.clipInput  = True

enc = ScalarEncoder(scalarEncoderParams)
```

```python
a = 1     #
fs = 8000 #
f0 = 440  #
sec = 0.1   #
 
swav=[]
 
for n in np.arange(fs * sec):
    #
    s = a * np.sin(2.0 * np.pi * f0 * n / fs)
    s_SDR=enc.encode(s)
    print(s)
    #print(s_SDR)
    swav.append(s)

 
#
plt.plot(swav[0:100])
plt.show()
```

```python
a = 1
fs = 8000
f0 = 440
sec = 0.1

predictions = {1: [], 2: []}
 
for n in np.arange(fs * sec):
    s = a * np.sin(2.0 * np.pi * f0 * n / fs)
    s_SDR=enc.encode(s)
    tm.compute(s_SDR, learn = True)
    tm.activateDendrites(True)
    #print(tm.getPredictiveCells())
    
    # .
    pdf = predictor.infer( tm.getActiveCells() )
    for n in (1,2):
        if pdf[n]:
            predictions[n].append( np.argmax( pdf[n] ) * predictor_resolution )
        else:
            predictions[n].append(float('nan'))
            
    predictor.learn(n, tm.getActiveCells(), int(s / predictor_resolution))
```

```python
predictor.reset()
A = predictor.infer( tm.getActiveCells())

print(A)
print(np.argmax( A[1] ) )
print(np.argmax( A[2] ) )
```
