## 簡単なMidi作成

```python
notes=[[60,1],[60,1],[67,1],[67,1],[69,1],[69,1],[67,2]]
```

```python
from music21 import *

stream2 = stream.Stream()

for n in range(len(notes)):
    noteMidi = notes[n][0]
    noteLength=notes[n][1]
    n = note.Note(midi=noteMidi,quarterLength = noteLength)

    stream2.append(n)
    
stream2.show()
StreamPlayer=midi.realtime.StreamPlayer(stream2)
StreamPlayer.play()
```

## 簡単なエンコード

```python
from music21 import *

inputMidi = converter.parse('twinkle_twinkle_PNOTR.mid')
parseMidi = inputMidi.parts[0].recurse()
notes = []
for e in parseMidi:
    if isinstance(e, note.Note):
        notes.append([int(e.pitch.ps),float(round(e.quarterLength,6))])
```

## 簡単な和音エンコード

```python
from music21 import *

inputMidi = converter.parse('twinkle_twinkle.mid')
parseMidi = inputMidi.parts[0].recurse()
notes = []
g=[]
for e in parseMidi:
    g=[]
    if isinstance(e, chord.Chord):
        for f in range(len(e.pitches)):
            g.append(int(e.pitches[f].ps))
        notes.append([g,float(round(e.quarterLength,6))])
    elif isinstance(e, note.Note):
        notes.append([int(e.pitch.ps),float(round(e.quarterLength,6))])
print(notes)
```

```python
from music21 import *

stream3 = stream.Stream()

for n in range(len(notes)):
    if type(notes[n][0]) ==list:
        chordMidi =chord.Chord(notes[n][0])
        noteLength=notes[n][1]
        n = chord.Chord(chordMidi,quarterLength = noteLength)
    else:
        noteMidi = notes[n][0]
        noteLength=notes[n][1]
        n = note.Note(midi=noteMidi,quarterLength = noteLength)

    stream3.append(n)
    
stream3.show()
StreamPlayer=midi.realtime.StreamPlayer(stream3)
StreamPlayer.play()
```
